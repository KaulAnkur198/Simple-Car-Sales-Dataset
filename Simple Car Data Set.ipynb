## Dealing with missing data

```python
cars_df_missing = pd.read_csv("Basic Data/car-sales-missing-data.csv")
cars_df_missing
```

```python
cars_df_missing.describe()
```

```python
cars_df_missing.info()
```

#### Count the missing values per column

```python
cars_df_missing.isnull().sum()
```

### Filling the missing values with mean in 'Odometer' column

```python
cars_df_missing_fill_values = cars_df_missing.copy()
```

```python
cars_df_missing_fill_values['Odometer'] = cars_df_missing['Odometer'].fillna(cars_df_missing['Odometer'].mean())
```

```python
cars_df_missing_fill_values
```

### Dropping the missing values row from the data

```python
cars_df_missing_fill_values.dropna(inplace=True)
cars_df_missing_fill_values
```

```python
cars_df_missing
```

```python
cars_df_missing.isnull().sum()
```

### Adding a new column

```python
no_of_seats = pd.Series([5,4,5,7,4])
```

```python
cars_df_missing['Seats'] = no_of_seats
cars_df_missing
```

```python
cars_df_missing['Seats'] = cars_df_missing['Seats'].fillna(5)
```

```python
cars_df_missing
```

```python
cars_df_missing['Seats'] = cars_df_missing['Seats'].astype(int)
cars_df_missing
```

```python
fuel_economy = [5.6,7,3,4.9,8.4,7,8,7.2,8,7]
cars_df_missing['Fuel Per 100 KM'] = fuel_economy
cars_df_missing
```

```python
cars_df_missing['Odometer'] = cars_df_missing['Odometer'].fillna(cars_df_missing['Odometer'].mean())
cars_df_missing
```

```python
cars_df_missing['Doors'] = cars_df_missing['Doors'].fillna(4)
cars_df_missing
```

```python
cars_df_missing['Doors'] = cars_df_missing['Doors'].fillna(4)
cars_df_missing.dropna(inplace=True)
```

```python
cars_df_missing
```

### Adding a new column from existing column

```python
cars_df_missing['Total Fuel Used'] = cars_df_missing['Odometer'] /100 * cars_df_missing['Fuel Per 100 KM']
cars_df_missing['Total Fuel Used']
```
